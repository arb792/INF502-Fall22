```python
import requests
from bs4 import BeautifulSoup
```

```python
result = requests.get("https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population_density")
```

```python
content = result.content
soup = BeautifulSoup(content, "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
for item in soup.find_all('a'):
    if (item.has_attr('title')):
        print("title is: " + item.get("title")+"\n")
        
```

```python
for item in soup.table.find_all('a'):
    if (item.has_attr('title')):
        print("title is: " + item.get("title")+"\n")
```

```python
for item in soup.table.find_all('td'):
    if (item.has_attr('td.text')):
            print("td.text is: "+ item.get("td.text")+"\n")
```
